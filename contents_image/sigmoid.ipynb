```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
values = np.arange(-10, 20, 0.1)
achievements = sigmoid(values)
pains = 1-sigmoid(values)

df = pd.DataFrame({'시간':np.arange(len(values)), 
                   '성취감':achievements,
                     '고통':pains})


fig = px.line(df, x='시간', y=['성취감', '고통']).update_layout(yaxis_title='크기')

fig.show()
```

```python
values = np.arange(-10, 20, 0.1)

plt.plot(values, sigmoid(values))
plt.plot(values, -1*sigmoid(values) +1)
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function in Matplotlib')
plt.show()
```
